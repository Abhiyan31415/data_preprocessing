# Neplish ASR Data Preprocessing Pipeline

This notebook preprocesses the raw audio/transcript data for our Nepali-English (Neplish) code-switched ASR project.

**Pipeline:**
1. Parse transcripts with timestamps
2. Segment audio files
3. Clean transcripts (resolve bracket annotations)
4. Create HuggingFace dataset with train/val/test splits
5. Apply data augmentation
6. Save all statistics for the presentation

```python
# Install required packages if needed
# !pip install pandas numpy librosa pydub soundfile datasets tqdm matplotlib seaborn
```

```python
import os
import re
import json
from pathlib import Path
from collections import Counter

import pandas as pd
import numpy as np
from tqdm.notebook import tqdm

# Audio processing
import librosa
import soundfile as sf
from pydub import AudioSegment

# HuggingFace datasets
from datasets import Dataset, DatasetDict, Audio

# Set paths
PROJECT_ROOT = Path(os.getcwd())
RAW_DIR = PROJECT_ROOT / "raw_dataset"
DATA_DIR = PROJECT_ROOT / "data"
REPORTS_DIR = PROJECT_ROOT / "reports"

# Create directories
DATA_DIR.mkdir(exist_ok=True)
(DATA_DIR / "segmented_audio").mkdir(exist_ok=True)
REPORTS_DIR.mkdir(exist_ok=True)
(REPORTS_DIR / "figures").mkdir(exist_ok=True)

print(f"Project root: {PROJECT_ROOT}")
print(f"Raw data: {RAW_DIR}")
print(f"Output: {DATA_DIR}")
```

## 1. Parse Transcripts

Parse the raw transcript files with timestamps `[MM:SS MM:SS]` format and extract segments.

```python
def parse_timestamp(ts_str: str) -> float:
    """Convert MM:SS or M:MM to seconds."""
    parts = ts_str.strip().split(":")
    if len(parts) == 2:
        minutes = int(parts[0])
        seconds = float(parts[1])
        return minutes * 60 + seconds
    return 0.0


def parse_transcript_file(txt_path: Path, audio_name: str) -> list:
    """Parse a single transcript file."""
    content = txt_path.read_text(encoding="utf-8")
    segments = []
    
    # Pattern for timestamps: [MM:SS MM:SS] or variations
    # Handles: [00:00 00:22], [0:17 - 0:36], [00:13 00:30], etc.
    pattern = r'\[(\d+:\d+)[\s\-]+(\d+:\d+)\]'
    
    lines = content.split('\n')
    current_speaker = audio_name  # Default speaker is the file name
    segment_id = 0
    
    for line in lines:
        line = line.strip()
        if not line:
            continue
            
        # Check for speaker label
        if line.startswith("Speaker"):
            current_speaker = line.strip().replace(" ", "_")
            continue
        
        # Look for timestamp pattern
        match = re.search(pattern, line)
        if match:
            start_ts = match.group(1)
            end_ts = match.group(2)
            
            # Extract text after the timestamp marker
            text_start = line.find('>') + 1 if '>' in line else match.end()
            raw_text = line[text_start:].strip()
            
            # Skip empty text
            if raw_text:
                segments.append({
                    "audio_file": audio_name,
                    "speaker": current_speaker,
                    "start_time": parse_timestamp(start_ts),
                    "end_time": parse_timestamp(end_ts),
                    "raw_text": raw_text,
                    "segment_id": f"{audio_name}_{segment_id:03d}",
                })
                segment_id += 1
    
    return segments


# Parse all transcript files
all_segments = []

for txt_file in sorted(RAW_DIR.glob("*.txt")):
    audio_name = txt_file.stem
    print(f"Parsing {txt_file.name}...")
    segments = parse_transcript_file(txt_file, audio_name)
    all_segments.extend(segments)
    print(f"  Found {len(segments)} segments")

parsed_df = pd.DataFrame(all_segments)
print(f"\nTotal segments parsed: {len(parsed_df)}")
parsed_df.head(10)
```

```python
# Save parsed transcripts
parsed_df.to_csv(DATA_DIR / "parsed_transcripts.csv", index=False)
print(f"Saved to: {DATA_DIR / 'parsed_transcripts.csv'}")

# Show statistics
print(f"\n{'='*50}")
print(f"Parsed Transcripts Summary")
print(f"{'='*50}")
print(f"  Total segments: {len(parsed_df)}")
print(f"  Unique speakers: {parsed_df['speaker'].nunique()}")
print(f"  Speakers: {parsed_df['speaker'].unique().tolist()}")
print(f"  Audio files: {parsed_df['audio_file'].unique().tolist()}")
```

## 2. Segment Audio Files

Extract audio segments based on timestamps. Convert to 16kHz mono WAV (Whisper-compatible).

```python
def segment_audio(audio_path: Path, segments_df: pd.DataFrame, output_dir: Path):
    """Segment a single audio file based on timestamps."""
    print(f"Loading {audio_path.name}...")
    
    # Load audio using pydub (handles AAC)
    audio = AudioSegment.from_file(str(audio_path))
    
    # Convert to 16kHz mono
    audio = audio.set_frame_rate(16000).set_channels(1)
    
    output_dir.mkdir(parents=True, exist_ok=True)
    
    durations = []
    for _, row in tqdm(segments_df.iterrows(), total=len(segments_df), desc="Segmenting"):
        start_ms = int(row['start_time'] * 1000)
        end_ms = int(row['end_time'] * 1000)
        
        segment = audio[start_ms:end_ms]
        
        output_path = output_dir / f"{row['segment_id']}.wav"
        segment.export(str(output_path), format="wav")
        
        duration = len(segment) / 1000
        durations.append(duration)
    
    return durations


# Segment all audio files
all_durations = []
segment_paths = []

for audio_file in parsed_df['audio_file'].unique():
    audio_path = RAW_DIR / f"{audio_file}.aac"
    if not audio_path.exists():
        print(f"Warning: {audio_path} not found, skipping...")
        continue
    
    file_segments = parsed_df[parsed_df['audio_file'] == audio_file]
    output_dir = DATA_DIR / "segmented_audio" / audio_file
    
    durations = segment_audio(audio_path, file_segments, output_dir)
    all_durations.extend(durations)
    
    for _, row in file_segments.iterrows():
        segment_paths.append(str(output_dir / f"{row['segment_id']}.wav"))

# Add duration and path to dataframe
parsed_df['duration'] = all_durations
parsed_df['audio_path'] = segment_paths

print(f"\n{'='*50}")
print(f"Audio Segmentation Complete")
print(f"{'='*50}")
print(f"  Total segments: {len(all_durations)}")
print(f"  Total duration: {sum(all_durations)/60:.2f} minutes")
print(f"  Mean duration: {np.mean(all_durations):.2f} seconds")
print(f"  Min/Max duration: {min(all_durations):.2f}s / {max(all_durations):.2f}s")
```

## 3. Clean Transcripts

Apply cleaning rules to resolve bracket annotations (following minor-testing patterns):
- `[English/Nepali]` -> Keep English word
- `[<unk>]`, `[<sil>]`, `[NOISE]` -> Remove
- `[word/<unk>]` -> Keep the word
- etc.

```python
def clean_transcript(text: str, keep_english: bool = True) -> str:
    """Clean a single transcript string by resolving all bracket annotations."""
    if not isinstance(text, str) or not text.strip():
        return ""

    # 1. [<fin>] -> remove (end-of-utterance metadata)
    text = re.sub(r"\[<fin>\]", "", text)

    # 2. [<sil>] -> remove (silence)
    text = re.sub(r"\[<sil>\]", "", text)

    # 3. [<unk>] standalone -> remove (unintelligible)
    text = re.sub(r"\[<unk>\]", "", text)

    # 4. [NOISE], [noise], [silence] -> remove
    text = re.sub(r"\[NOISE\]", "", text, flags=re.IGNORECASE)
    text = re.sub(r"\[noise\]", "", text, flags=re.IGNORECASE)
    text = re.sub(r"\[silence\]", "", text, flags=re.IGNORECASE)

    # 5. [<rep>/word] -> keep the word
    text = re.sub(r"\[<rep>/([^\]]+)\]", r"\1", text)

    # 6. [<fil>/word] -> keep the word
    text = re.sub(r"\[<fil>/([^\]]+)\]", r"\1", text)

    # 7. [<unk>/word] -> keep the word
    text = re.sub(r"\[<unk>/([^\]]+)\]", r"\1", text)

    # 8. [word/<unk>] -> keep the word
    text = re.sub(r"\[([^\]<]+)/<unk>\]", r"\1", text)

    # 9. [English/Nepali] code-switch markers
    def _resolve_code_switch(match):
        part1 = match.group(1).strip()
        part2 = match.group(2).strip()
        # Check if part1 is Latin-script (English)
        if re.fullmatch(r"[a-zA-Z0-9\s\-'\.]+", part1):
            return part1 if keep_english else part2
        return part1

    text = re.sub(r"\[([^\]]+)/([^\]]+)\]", _resolve_code_switch, text)

    # 10. Remove any remaining bare special tokens
    text = re.sub(r"<(?:unk|sil|rep|fin|fil)>", "", text)

    # 11. Remove inline timestamps like [00:13]
    text = re.sub(r"\[\d+:\d+\]", "", text)

    # 12. Normalize whitespace
    text = re.sub(r"\s+", " ", text).strip()

    return text


# Apply cleaning
print("Cleaning transcripts...")
parsed_df['sentence'] = parsed_df['raw_text'].apply(clean_transcript)

# Track cleaning statistics
parsed_df['raw_char_len'] = parsed_df['raw_text'].str.len()
parsed_df['clean_char_len'] = parsed_df['sentence'].str.len()
parsed_df['chars_removed'] = parsed_df['raw_char_len'] - parsed_df['clean_char_len']

# Check for code-switching
parsed_df['has_english'] = parsed_df['sentence'].apply(
    lambda s: bool(re.search(r'[a-zA-Z]{2,}', str(s)))
)

# Drop empty sentences
before_count = len(parsed_df)
cleaned_df = parsed_df[parsed_df['sentence'].str.len() > 0].reset_index(drop=True)
dropped_count = before_count - len(cleaned_df)

print(f"\n{'='*50}")
print(f"Cleaning Summary")
print(f"{'='*50}")
print(f"  Total samples kept: {len(cleaned_df)}")
print(f"  Dropped (empty): {dropped_count}")
print(f"  Code-switched: {cleaned_df['has_english'].sum()} ({cleaned_df['has_english'].mean():.1%})")
print(f"  Pure Nepali: {(~cleaned_df['has_english']).sum()} ({(~cleaned_df['has_english']).mean():.1%})")
print(f"  Avg chars removed: {cleaned_df['chars_removed'].mean():.1f}")
print(f"  Avg sentence length: {cleaned_df['clean_char_len'].mean():.1f} chars")
```

## 4. Create HuggingFace Dataset

Create a HuggingFace dataset with train/validation/test splits.

```python
def create_hf_dataset(df: pd.DataFrame, audio_col: str = 'audio_path', 
                       text_col: str = 'sentence') -> Dataset:
    """Create a HuggingFace Dataset from a DataFrame."""
    # Prepare data dictionary
    data = {
        'audio': df[audio_col].tolist(),
        'sentence': df[text_col].tolist(),
        'speaker': df['speaker'].tolist(),
        'segment_id': df['segment_id'].tolist(),
        'duration': df['duration'].tolist(),
        'has_english': df['has_english'].tolist(),
    }
    
    # Create dataset
    dataset = Dataset.from_dict(data)
    
    # Cast audio column to Audio feature (16kHz)
    dataset = dataset.cast_column('audio', Audio(sampling_rate=16000))
    
    return dataset


# Create the dataset
print("Creating HuggingFace dataset...")
full_dataset = create_hf_dataset(cleaned_df)
print(f"Dataset created with {len(full_dataset)} samples")
print(f"Features: {full_dataset.features}")
```

```python
# Create train/val/test splits
# Using 80/10/10 split
from sklearn.model_selection import train_test_split

# First split: 80% train, 20% temp
train_indices, temp_indices = train_test_split(
    range(len(full_dataset)), 
    test_size=0.2, 
    random_state=42,
    stratify=cleaned_df['speaker'].tolist()  # Stratify by speaker
)

# Second split: 50% val, 50% test from temp
temp_speakers = [cleaned_df.iloc[i]['speaker'] for i in temp_indices]
val_indices, test_indices = train_test_split(
    temp_indices,
    test_size=0.5,
    random_state=42,
    stratify=temp_speakers
)

# Create DatasetDict
dataset_dict = DatasetDict({
    'train': full_dataset.select(train_indices),
    'validation': full_dataset.select(val_indices),
    'test': full_dataset.select(test_indices),
})

print(f"\n{'='*50}")
print(f"Dataset Splits")
print(f"{'='*50}")
for split_name, split_ds in dataset_dict.items():
    print(f"  {split_name}: {len(split_ds)} samples")

# Save the dataset
hf_dataset_path = DATA_DIR / "hf_dataset"
dataset_dict.save_to_disk(str(hf_dataset_path))
print(f"\nSaved to: {hf_dataset_path}")
```

## 5. Data Augmentation

Apply conservative audio augmentations to the training set:
- Gaussian noise (SNR 20-40 dB)
- Tempo change (0.93x - 1.07x)
- Pitch shifting (±1.5 semitones)
- Volume perturbation (±4 dB)

```python
# Augmentation functions
def add_gaussian_noise(audio: np.ndarray, rng: np.random.Generator, 
                        snr_range: tuple = (20.0, 40.0)) -> np.ndarray:
    """Add Gaussian white noise at a random SNR."""
    snr_db = rng.uniform(*snr_range)
    audio_power = np.mean(audio ** 2)
    if audio_power < 1e-10:
        return audio
    
    noise = rng.standard_normal(len(audio)).astype(np.float32)
    noise_power = np.mean(noise ** 2)
    target_noise_power = audio_power / (10 ** (snr_db / 10))
    scale = np.sqrt(target_noise_power / max(noise_power, 1e-10))
    
    result = (audio + scale * noise).astype(np.float32)
    return np.clip(result, -1.0, 1.0)


def change_tempo(audio: np.ndarray, sr: int, rng: np.random.Generator,
                  speed_range: tuple = (0.93, 1.07)) -> np.ndarray:
    """Change tempo without changing pitch using librosa.effects.time_stretch."""
    rate = rng.uniform(*speed_range)
    augmented = librosa.effects.time_stretch(y=audio, rate=rate)
    
    # Trim to 30s if needed
    max_len = int(30.0 * sr)
    if len(augmented) > max_len:
        augmented = augmented[:max_len]
    
    return augmented.astype(np.float32)


def shift_pitch(audio: np.ndarray, sr: int, rng: np.random.Generator,
                 semitone_range: tuple = (-1.5, 1.5)) -> np.ndarray:
    """Shift pitch by a small random amount."""
    n_steps = rng.uniform(*semitone_range)
    if abs(n_steps) < 0.1:
        return audio
    
    augmented = librosa.effects.pitch_shift(y=audio, sr=sr, n_steps=n_steps)
    return np.clip(augmented, -1.0, 1.0).astype(np.float32)


def change_volume(audio: np.ndarray, rng: np.random.Generator,
                   gain_db_range: tuple = (-4.0, 4.0)) -> np.ndarray:
    """Apply random volume gain in dB."""
    gain_db = rng.uniform(*gain_db_range)
    gain_linear = 10 ** (gain_db / 20)
    augmented = audio * gain_linear
    return np.clip(augmented, -1.0, 1.0).astype(np.float32)


def augment_audio(audio: np.ndarray, sr: int, rng: np.random.Generator) -> tuple:
    """Apply exactly ONE random augmentation."""
    augmentations = ['noise', 'tempo', 'pitch', 'volume']
    choice = rng.choice(augmentations)
    
    if choice == 'noise':
        return add_gaussian_noise(audio, rng), choice
    elif choice == 'tempo':
        return change_tempo(audio, sr, rng), choice
    elif choice == 'pitch':
        return shift_pitch(audio, sr, rng), choice
    elif choice == 'volume':
        return change_volume(audio, rng), choice
    
    return audio, 'none'
```

```python
# Augment the training set
print("Augmenting training data...")

train_ds = dataset_dict['train']
augmentation_stats = Counter()

augmented_samples = []

rng = np.random.default_rng(42)

for i, example in enumerate(tqdm(train_ds, desc="Augmenting")):
    audio_data = example['audio']
    audio_array = np.array(audio_data['array'], dtype=np.float32)
    sr = audio_data['sampling_rate']
    
    # Create one augmented copy
    aug_audio, aug_type = augment_audio(audio_array, sr, rng)
    augmentation_stats[aug_type] += 1
    
    augmented_samples.append({
        'audio': {'array': aug_audio, 'sampling_rate': sr},
        'sentence': example['sentence'],
        'speaker': example['speaker'],
        'segment_id': f"{example['segment_id']}_aug",
        'duration': len(aug_audio) / sr,
        'has_english': example['has_english'],
        'augmentation_type': aug_type,
    })

print(f"\nAugmentation types applied:")
for aug_type, count in augmentation_stats.items():
    print(f"  {aug_type}: {count}")
```

```python
# Create augmented dataset
augmented_ds = Dataset.from_list(augmented_samples)
augmented_ds = augmented_ds.cast_column('audio', Audio(sampling_rate=16000))

# Add augmentation_type column to original train set (as 'original')
train_with_type = train_ds.add_column('augmentation_type', ['original'] * len(train_ds))

# Concatenate original + augmented
from datasets import concatenate_datasets

augmented_train = concatenate_datasets([train_with_type, augmented_ds])
augmented_train = augmented_train.shuffle(seed=42)

# Create final augmented dataset dict
augmented_dataset_dict = DatasetDict({
    'train': augmented_train,
    'validation': dataset_dict['validation'],
    'test': dataset_dict['test'],
})

print(f"\n{'='*50}")
print(f"Augmentation Summary")
print(f"{'='*50}")
print(f"  Original train: {len(train_ds)} samples")
print(f"  Augmented train: {len(augmented_train)} samples")
print(f"  Growth factor: {len(augmented_train) / len(train_ds):.2f}x")

# Save augmented dataset
augmented_path = DATA_DIR / "hf_dataset_augmented"
augmented_dataset_dict.save_to_disk(str(augmented_path))
print(f"\nSaved to: {augmented_path}")
```

## 6. Save Statistics for Presentation

Save all computed statistics to JSON and markdown files.

```python
# Compute comprehensive statistics
durations = np.array(cleaned_df['duration'].tolist())

# Extract English words for code-switching analysis
all_english_words = []
for sent in cleaned_df['sentence']:
    all_english_words.extend(re.findall(r'[a-zA-Z]{2,}', str(sent)))

word_counts = Counter(all_english_words)
top_20_english = word_counts.most_common(20)

# Build statistics dictionary
statistics = {
    "raw_data": {
        "num_audio_files": len(parsed_df['audio_file'].unique()),
        "audio_files": parsed_df['audio_file'].unique().tolist(),
        "num_speakers": len(parsed_df['speaker'].unique()),
        "speakers": parsed_df['speaker'].unique().tolist(),
    },
    "audio": {
        "total_segments": len(cleaned_df),
        "total_duration_seconds": float(durations.sum()),
        "total_duration_minutes": float(durations.sum() / 60),
        "mean_duration": float(durations.mean()),
        "median_duration": float(np.median(durations)),
        "min_duration": float(durations.min()),
        "max_duration": float(durations.max()),
        "std_duration": float(durations.std()),
    },
    "transcript": {
        "total_samples": len(cleaned_df),
        "total_characters_raw": int(cleaned_df['raw_char_len'].sum()),
        "total_characters_clean": int(cleaned_df['clean_char_len'].sum()),
        "total_chars_removed": int(cleaned_df['chars_removed'].sum()),
        "mean_sentence_length": float(cleaned_df['clean_char_len'].mean()),
        "median_sentence_length": float(cleaned_df['clean_char_len'].median()),
    },
    "code_switching": {
        "code_switched_count": int(cleaned_df['has_english'].sum()),
        "code_switched_pct": float(cleaned_df['has_english'].mean() * 100),
        "pure_nepali_count": int((~cleaned_df['has_english']).sum()),
        "pure_nepali_pct": float((~cleaned_df['has_english']).mean() * 100),
        "unique_english_words": len(set(all_english_words)),
        "total_english_occurrences": len(all_english_words),
        "top_20_english_words": top_20_english,
    },
    "splits": {
        "train": len(dataset_dict['train']),
        "validation": len(dataset_dict['validation']),
        "test": len(dataset_dict['test']),
    },
    "augmentation": {
        "original_train_size": len(dataset_dict['train']),
        "augmented_train_size": len(augmented_dataset_dict['train']),
        "growth_factor": len(augmented_dataset_dict['train']) / len(dataset_dict['train']),
        "augmentation_types": dict(augmentation_stats),
    },
    "speaker_distribution": cleaned_df['speaker'].value_counts().to_dict(),
}

# Save to JSON
with open(REPORTS_DIR / "dataset_statistics.json", 'w', encoding='utf-8') as f:
    json.dump(statistics, f, indent=2, ensure_ascii=False)

print(f"Saved statistics to: {REPORTS_DIR / 'dataset_statistics.json'}")
```
